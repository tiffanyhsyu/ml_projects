```python
# Run on all systems
scale, width, height = 0.05, 40, 40

for i in range(len(selected)):
    name = 'J' + selected[i]['col1'].replace(':','')[0:4] + selected[i]['col2'].replace(':','')[0] + \
                    selected[i]['col2'].replace(':','')[1:5]
    print (i, 'Working on ', name)
    coords = SkyCoord(selected[i]['col1'], selected[i]['col2'], unit=(u.hourangle, u.deg))
    url = ('http://skyserver.sdss.org/dr16/SkyServerWS/ImgCutout/getjpeg?TaskName=Skyserver.QuickLook.Image&ra=%.8f&dec=%.8f&scale=%.2f&width=%i&height=%i'
                    % (coords.ra.value, coords.dec.value, scale, width, height))
    print ('\t Downloading %s' % url)
    
    img_data = requests.get(url).content
    with open('/Users/thsyu/Google Drive File Stream/My Drive/BCDs/talks/2020_05_defense/figures/downloads/' + name + '.jpg', 'wb') as handler:
        print ('\t Saving image')
        handler.write(img_data)
```
